```python
# Solicite al usuario que introduzca 3 números enteros separados por comas (por ejemplo: 1,2,3).

entrada = input("Porfavor introduzca 3 números enteros separados por comas 'Por ejemplo: 1,2,3'")

# Convertir la entrada en una lista de enteros

numeros = [int(x) for x in entrada.split(",")]

# Mostrar los números para comprobar

print("Los numeros introducidos son:", numeros)

# Utilizando funciones built-in, determine el número máximo, el número mínimo y la suma total de los números introducidos.  

print("El numero maximo es:\n", max(numeros),
      "\nEl numero minimo es:\n", min(numeros),
      "\nY la suma total de los numeros es:", sum(numeros))

# Determine la verdad o falsedad de la siguiente afirmación: "El número máximo menos el número mínimo es igual a la suma total dividida por 5".

afirmacion = max(numeros) - min(numeros) == sum(numeros) / 5
print("Por tanto la afirmacion es:", afirmacion)
```
